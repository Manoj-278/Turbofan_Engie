In the education column of our dataset, we have identified the presence of 14 missing values. Given that this column is categorical, employing the mode for data imputation is a suitable approach. By replacing the missing values with the mode, we aim to maintain the integrity of the categorical distribution within the education feature, ensuring a robust and representative dataset for our analysis

```python

df1['EDUCATION'].fillna(df1['EDUCATION'].mode().iloc[0],inplace=True)
```

```python
df1.columns
```

If you use the same numbers to mean different things in different categories, it might confuse your model. For instance, if '1' means 'married' in one field and 'graduate school' in another, the model might mix them up. This could lead to inaccurate results when testing ideas. To avoid this, it's important to make sure that each number has a clear and unique meaning in its context. This helps the model understand the differences correctly and gives you more reliable results in hypothesis testing.

```python
df1['GENDER'].replace([1,2],['Male','Female'],inplace=True)
df1['EDUCATION'].replace([1,2,3,4,5],['graduate school','university','high school','others','unknown'],inplace=True)
df1['MARITAL STATUS'].replace([1,2,3],['married','single','others'],inplace=True)
```

```python
df1['EDUCATION'].value_counts()
```

```python
# As are having few unknown values in education col we can perform data imputation
df1['EDUCATION'].replace('unknown',df1['EDUCATION'].mode()[0],inplace=True)
```

In addressing the presence of outliers in our dataset, we have opted to employ the Z-score method for outlier detection. Setting a threshold of 3, any data points with a Z-score beyond this limit will be identified and treated as outliers. This approach ensures that we capture and appropriately handle data points that significantly deviate from the mean, contributing to a more robust and accurate analysis of our dataset

```python
import scipy.stats as stats

# Calculate Z-scores for the specified numeric columns
z = np.abs(stats.zscore(df1[num_col]))

# df2 by keeping only rows where all Z-scores are less than 3
# common threshold for identifying outliers, where Z-score > 3 is considered an outlier
df2 = df1[(z<3).all(axis=1)]

# see the shape to find out how much outlier we removed

print(df1.shape)
print(df2.shape)
```

```python
cat_col= df2.select_dtypes(include=object)
cat_col
```

 Below code is creating a series of bar plots, each showing the percentage of defaulters for different categories within a categorical column. It provides a visual representation of how default rates vary across different categories

```python
for i in cat_col:
    plt.figure(figsize=(8, 6))

    # Calculate percentage of defaulters for each category
    percentages = df2.groupby(i)['DEF_AMT'].mean() * 100

    # Plot the bar chart
    sns.barplot(x=percentages.index, y=percentages.values)

    plt.title(f'Percentage of Defaulters for {i}')
    plt.xlabel(i)
    plt.ylabel('Percentage')
    plt.show()
```

Histograms are excellent visual tools for understanding the distribution of continuous variables. In a right-skewed distribution, the histogram would typically show a tail extending to the right (positively skewed), indicating the presence of larger values on that side.

By observing the histogram:

### Mode :
The peak of the histogram would be on the left side, representing the most frequently occurring values.
### Median:
The median would be located closer to the left side of the distribution, where the bulk of the data is concentrated.
### Mean:
The mean would be influenced by the presence of larger values in the right tail, pulling it in that direction.

```python
sns.set(style="whitegrid")

num_rows = len(num_col) // 3 + (len(num_col) % 3 > 0)

# Create subplots for each continuous variable
fig, axes = plt.subplots(nrows=num_rows, ncols=3, figsize=(15, 5 * 6))

axes = axes.flatten()

for i, col in enumerate(num_col):
    sns.histplot(df2[col], kde=True, bins=10, color='skyblue', stat='count', ax=axes[i])
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Density')
    axes[i].set_title(f'Histogram with Density Plot for {col}')

# Remove empty subplots if the number of variables is not a multiple of 3
for i in range(len(num_col), len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.show()
```
